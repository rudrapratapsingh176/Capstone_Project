# 2.5) Analysis of population vs overall crime!!

```python
plt.figure(figsize=(15,8),facecolor='Yellow')
sns.scatterplot(x=df['Population'],y=df['Total'])
plt.xlabel("TOTAL POPULATION")
plt.ylabel("OVERALL CRIEMES!!")
plt.title("OVERALL CRIMES VS THE TOTAL POPULATION")
plt.grid()
plt.show()
```

The relationship between population growth and crime rates is a complex issue with many factors that contribute to it. While there is some evidence to suggest that there may be a correlation between population growth and crime rates, it is not a direct relationship.

One reason why an increase in population could lead to an increase in crime is the concept of strain theory. This theory suggests that when people experience strains such as poverty, unemployment, or lack of resources, they may become frustrated and turn to criminal behavior as a means of coping. When a population grows, there may be more competition for resources and opportunities, leading to more strains and potentially more criminal behavior.

Additionally, as the population increases, so does the diversity of the community. Diversity can lead to cultural clashes and conflicts, which may contribute to an increase in crime. Moreover, a larger population can lead to a breakdown in social networks and community ties, making it easier for criminals to go unnoticed and commit crimes.
